```python
from pprint import pprint

def print_response(response):
  print(f"ResponceId: {response.id}")
  print(f"Input tokens: {response.usage.input_tokens} ({response.usage.input_tokens_details.cached_tokens} cched) | Output tokens: {response.usage.output_tokens} ({response.usage.output_tokens_details.reasoning_tokens} reasoning)")
  pprint(response.output)

  print()
  print(f"{'='*20} [Text] {'='*20}")
  print(response.text)
  print()
  print(f"{'='*20} [Finish Reason] {'='*20}")
  pprint(response.output_text)
```

```python
api_Key = userdata.get("AIKey")
```

```python
client = OpenAI(api_key=api_Key)

response = client.responses.create(
    model="gpt-5-nano",
    input="Write a short bedtime story about a unicorn.",
    instructions="exactly 3 sentencises, max 50 words"
)

```

```python
print_response(response)
```

```python
second_response = client.responses.create(
    model="gpt-5-nano",
    input="The unicorn meets a bear and they go for walk. What happens next?",
    previous_response_id=response.id
)
```

```python
print_response(second_response)
```

```python
with client.responses.stream(
    model="gpt-5-nano",
    input="Tell me more about RAG in contect of AI",
    instructions="Short answer 3 sentances on ENG and 3 on BG"
)as stream:
  for event in stream:
    if event.type == 'response.output_text.delta':
      print(event.delta, end='', flush=True)
    print()

  stream_response = stream.get_final_response()
```

```python
print_response(stream_response)
```

```python
from pydantic import BaseModel

class Furniture(BaseModel):
  color: str
  width: int
  length: int
  height: int
  material: str
  name: str
  description: str

structured_response = client.responses.parse(
    model="gpt-5-nano",
    input="Generate a futuristic furniture that I can use in my IT office.",
    text_format=Furniture
)
```

```python
print_response(structured_response)
```
